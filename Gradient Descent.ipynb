# Derivatives & Slope

```python
def df(x):
    return 4 * x + 4
```

```python
#Plots
plt.figure(figsize = [15, 5])

# First Plot Cost Function
plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("f(x)", fontsize = 15)
plt.xlim(-10, 10)
plt.ylim(0, 200)
plt.grid()
plt.plot(x_1, f(x_1), color = "green", linewidth = 3)

# Second Plot Derivative
plt.subplot(1, 2, 2)
plt.title("Derivative", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("df(x)", fontsize = 15)
plt.grid()
plt.plot(x_1, df(x_1), color = "blue", linewidth = 3)

plt.show()
```

```python
# Gradient Descent Algorithm
current_X = 5
step = 0.1
X_list = []
for n in range(500000000000):
    gradient = df(current_X)
    # Append values to the lists
    X_list.append(current_X)
    # Stop when reach certain precision
    if abs(step * gradient) <= 0.0000001:
        print("The loop ran", n, "times.")
        break
    current_X = current_X - step * gradient
    

# Print out results
print("Local min occurs when X =", current_X)
print("The derivative at this point =", df(current_X))
print("f(x) value or cost at this point=", f(current_X))
```

```python
#Plots
plt.figure(figsize = [15, 12])

# Cost Function
plt.subplot(2, 2, 1)
plt.title("Cost Function", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("f(x)", fontsize = 15)
plt.xlim(-10, 10)
plt.ylim(0, 200)
plt.grid()
plt.plot(x_1, f(x_1), color = "green", linewidth = 3, alpha = 0.5)
x_array = np.array(X_list) 
plt.scatter(x_array, f(x_array), s = 100, alpha = 0.6)

# Cost Function (close up)
plt.subplot(2, 2, 2)
plt.title("Cost Function (close up)", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("f(x)", fontsize = 15)
plt.xlim(-1.2, 0)
plt.ylim(7, 10)
plt.grid()
plt.plot(x_1, f(x_1), color = "green", linewidth = 3, alpha = 0.5)
x_array = np.array(X_list) 
plt.scatter(x_array, f(x_array), s = 100, alpha = 0.6, color = "#ffc93c")

# Derivative
plt.subplot(2, 2, 3)
plt.title("Derivative", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("df(x)", fontsize = 15)
plt.grid()
plt.plot(x_1, df(x_1), color = "blue", linewidth = 3, alpha = 0.5)
plt.scatter(x_array, df(x_array), s = 100, alpha = 0.6, color = "green")

# Derivative (close up)
plt.subplot(2, 2, 4)
plt.title("Derivative (close up)", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("df(x)", fontsize = 15)
plt.xlim(-1.25, -0.5)
plt.ylim(-1, 2)
plt.grid()
plt.plot(x_1, df(x_1), color = "blue", linewidth = 3, alpha = 0.5)
plt.scatter(x_array, df(x_array), s = 100, alpha = 0.6, color = "#ffc93c")

plt.show()
```

# Example 2 - Multiple Minima vs Initial Guess & Advanced Functions
# $$g(x) = x^4 - 4x^2 + 5$$

```python
#Data Creation
x_2 = np.linspace(-2, 2, 1000)

#g(x)
def g(x):
    return x**4 - 4 * x**2 + 5

#dg(x)
def dg(x):
    return 4 * x**3 - 8 * x
```

```python
#Plots
plt.figure(figsize = [15, 5])

# First Plot Cost Function
plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("g(x)", fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(0, 6)
plt.grid()
plt.plot(x_2, g(x_2), color = "green", linewidth = 3)

# Second Plot Derivative
plt.subplot(1, 2, 2)
plt.title("Derivative", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("dg(x)", fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(-15, 15)
plt.grid()
plt.plot(x_2, dg(x_2), color = "blue", linewidth = 3)

plt.show()
```

```python
# Gradient Descent Function
# Batch gradient descent, has potential problems since it only finds the local minima, result varies depending on the chosen initial guess.
def gradient_descent_function(derivative_function, initial_guess, multiplier = 0.02, precision = 0.000001, max_iteration = 500):
    current_X = initial_guess
    X_list = []
    for n in range(max_iteration):
        gradient = derivative_function(current_X)
        # Append values to the lists
        X_list.append(current_X)
        # Stop when reach certain precision
        if abs(multiplier * gradient) <= precision:
            break
        current_X = current_X - multiplier * gradient
    return X_list
```

```python
# Plots
X_list2 = gradient_descent_function(dg, -0.5)
x_array2 = np.array(X_list2)

plt.figure(figsize = [15, 5])

# First Plot Cost Function
plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("g(x)", fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(0, 6)
plt.grid()
plt.plot(x_2, g(x_2), color = "blue", linewidth = 3)
plt.scatter(x_array2, g(x_array2), s = 100, alpha = 0.6, color = "green")

# Second Plot Derivative
plt.subplot(1, 2, 2)
plt.title("Derivative", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("dg(x)", fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(-15, 15)
plt.grid()
plt.plot(x_2, dg(x_2), color = "blue", linewidth = 3)
plt.scatter(x_array2, dg(x_array2), s = 100, alpha = 0.6, color = "green")

plt.show()
```

# Example 3 - Divergence and Overflow
## $$ h(x) = x^5 - 2x^4 + 2$$

```python
# Data Creation
x_3 = np.linspace(start = -2.5, stop = 2.5, num = 1000)

def h(x):
    return x**5 - 2 * x**4 + 2
def dh(x):
    return 5 * x**4 - 8 * x**3
```

```python
# Applied the previous batch gradient descent function.
# This will cause overflow if initial guess value is chosen to be outside of a certain range.
# In this case, 72 is the max iterations it can perform until it reaches the floating value limitation.
X_list3 = gradient_descent_function(dh, -0.2, max_iteration = 72)

# Convert list to array
x_array3 = np.array(X_list3)

# Calculate other parameters
local_min = x_array3[-1]
cost = h(local_min)
num_steps = len(x_array3)

# Plots
plt.figure(figsize = [15, 5])

# First Plot Cost Function
plt.subplot(1, 2, 1)
plt.title("Cost Function", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("h(x)", fontsize = 15)
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.grid()
plt.plot(x_3, h(x_3), color = "blue", linewidth = 3)
plt.scatter(x_array3, h(x_array3), s = 100, alpha = 0.6, color = "green")

# Second Plot Derivative
plt.subplot(1, 2, 2)
plt.title("Derivative", fontsize = 15)
plt.xlabel("X", fontsize = 15)
plt.ylabel("dh(x)", fontsize = 15)
plt.xlim(-2, 2)
plt.ylim(-4, 5)
plt.grid()
plt.plot(x_3, dh(x_3), color = "blue", linewidth = 3)
plt.scatter(x_array3, dh(x_array3), s = 100, alpha = 0.6, color = "green")

plt.show()
print("X:", local_min)
print("Cost:", cost)
print("Numer of steps:", num_steps)
print("This will cause overflow if the programme continues iterating!")
```

# Example 4 - Data Visualization with 3D Charts
## Minimize $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$
### Or $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
# Create the function
def f(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)
```

```python
# Data creation
x_4 = np.linspace(start = -2, stop = 2, num = 200)
y_4 = np.linspace(start = -2, stop = 2, num = 200)

# Convert to 2D arrays
x_4, y_4 = np.meshgrid(x_4, y_4)
```

```python
# Generating 3D Plot
fig = plt.figure(figsize = [14, 11])
plot3D = fig.gca(projection = "3d")
plot3D.set_xlabel("X", fontsize = 20)
plot3D.set_ylabel("Y", fontsize = 20)
plot3D.set_zlabel("f(x, y) -- Cost", fontsize = 20)
plot3D.plot_surface(x_4, y_4, f(x_4, y_4), cmap = cm.Spectral, alpha = 0.5)
plt.show()
```

# Partial Derivatives and Symbolic Computation
## $$\frac{\partial f}{\partial x} = \frac{2x \ln(3) \cdot 3^{-x^2 - y^2}}{\left (3^{-x^2 - y^2} + 1\right)^2}$$
## $$\frac{\partial f}{\partial y} = \frac{2y \ln(3) \cdot 3^{-x^2 - y^2}}{\left (3^{-x^2 - y^2} + 1\right)^2}$$

```python
a, b = symbols('x, y')
print("Cost function:",f(a, b))
print("Partial derevative with respect to x is:", diff(f(a, b), a))
print("Value of f(x, y) at x = 1.8 and y = 1.0 is:", 
      f(a, b).evalf(subs = {a : 1.8, b : 1.0}))
print("Value of the partial derivative with respect to x at x = 1.8 and y = 1.0 is:", 
      diff(f(a, b), a).evalf(subs = {a : 1.8, b : 1.0}))
```

# Batch Gradient Descent with Sympy

```python
# Use Sympy
# Set up initial values
multiplier = 0.1
max_iteration = 1000
# Initial guess
parameters = np.array([1.8, 1.0])
for n in range(max_iteration):
    gradient_x = diff(f(a, b), a).evalf(subs = {a : parameters[0], b: parameters[1]})
    gradient_y = diff(f(a, b), b).evalf(subs = {a : parameters[0], b: parameters[1]})
    gradients = np.array([gradient_x, gradient_y])
    parameters = parameters - multiplier * gradients
# Print results
print("Final gradients (x, y):", gradients)
print("Minimin cost occurs when x =", parameters[0])
print("Minimin cost occurs when y =", parameters[1])
print("Minimin cost =", f(parameters[0], parameters[1]))
print("Iteration:", n)
```

```python
# Implement partial derivatives manually
def fpx(x, y):
    return (2 * x * log(3) * 3**(-x**2 - y**2)) / ((3**(-x**2 - y**2) + 1)**2)
def fpy(x, y):
    return (2 * y * log(3) * 3**(-x**2 - y**2)) / ((3**(-x**2 - y**2) + 1)**2)
# Set up initial values
multiplier = 0.1
max_iteration = 1000
# Initial guess
parameters = np.array([1.8, 1.0])
for n in range(max_iteration):
    gradient_x = fpx(parameters[0], parameters[1])
    gradient_y = fpy(parameters[0], parameters[1])
    gradients = np.array([gradient_x, gradient_y])
    parameters = parameters - multiplier * gradients
# Print results
print("Final gradients (x, y):", gradients)
print("Minimin cost occurs when x =", parameters[0])
print("Minimin cost occurs when y =", parameters[1])
print("Minimin cost =", f(parameters[0], parameters[1]))
print("Iteration:", n)
```

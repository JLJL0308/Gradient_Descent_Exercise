# Import packages

```python
import matplotlib.pyplot as plt
import numpy as np

#Output the plot
%matplotlib inline
```

# Example 1: A simple cost function
## $f(x) = 2x^2 + 4x + 10$

```python
def f(x):
    return 2 * x**2 + 4 * x + 10
```

```python
# Make Data
x_1 = np.linspace(start = -10, stop = 10, num = 100)
```

```python
plt.xlabel("X", fontsize = 15)
plt.ylabel("f(x)", fontsize = 15)
plt.xlim(-10, 10)
plt.ylim(0, 200)
plt.plot(x_1, f(x_1))
plt.show()
```

# Derivatives & Slope

```python
def df(x):
    return 4 * x + 4
```

```python
plt.figure(figsize = [15, 5])

# First Plot Cost Function
plt.subplot(1, 2, 1)
plt.xlabel("X", fontsize = 15)
plt.ylabel("f(x)", fontsize = 15)
plt.xlim(-10, 10)
plt.ylim(0, 200)
plt.grid()
plt.plot(x_1, f(x_1), color = "green", linewidth = 4)

# Second Plot Derivative
plt.subplot(1, 2, 2)
plt.xlabel("X", fontsize = 15)
plt.ylabel("df(x)", fontsize = 15)
plt.grid()
plt.plot(x_1, df(x_1), color = "blue", linewidth = 4)

plt.show()
```
